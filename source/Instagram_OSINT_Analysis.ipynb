```python
# 3. Function to calculate engagement rate
def calculate_engagement_rate(likes, comments, followers):
    avg_likes = sum(likes) / len(likes) if likes else 0
    avg_comments = sum(comments) / len(comments) if comments else 0
    return ((avg_likes + avg_comments) / followers) * 100
```

```python
# 4. Function to plot the engagement rate using Plotly
def plot_engagement_rate(engagement_rate):
    fig = go.Figure(go.Indicator(
        mode="gauge+number",
        value=engagement_rate,
        title={"text": "Engagement Rate (%)"},
        gauge={"axis": {"range": [0, 100]}, "bar": {"color": "blue"}}
    ))
    fig.show(renderer="browser")
```

```python
# 5. Function to plot the relevant terms in the bio (Influencer-related keywords)
def plot_relevant_terms(relevant_terms):
    # Update the list of influencer-related keywords
    influencer_keywords = ["influencer", "content creator", "fashion", "travel", "beauty", "lifestyle", "fitness", "blogger", "photography", "fashionista", "makeup", "vlogger", "health", "sponsored", "collab", "advertising", "endorsements", "branding", "social media", "youtube", "tiktok", "digital marketer", "entrepreneur", "startup"]

    # Count the frequency of influencer-related keywords in the bio
    df_terms = pd.DataFrame({'Term': list(relevant_terms), 'Count': [1] * len(relevant_terms)})

    fig = px.bar(
        df_terms,
        x='Term',
        y='Count',
        title="Bio Keyword Analysis (Influencer-Related)",
        labels={'Term': 'Relevant Terms in Bio', 'Count': 'Frequency'},
        category_orders={'Term': influencer_keywords}  # Optional: to maintain keyword order
    )
    fig.update_yaxes(visible=False)  # Hide Y-axis values as they are all 1
    fig.show(renderer="browser")
```

```python
# 6. Main function for Instagram analysis
def instagram_osint_analysis(username):
    # Initialize Instaloader and fetch profile
    loader = instaloader.Instaloader()
    profile = instaloader.Profile.from_username(loader.context, username)

    # Basic profile info
    print(f"Analyzing @{username}")
    bio = profile.biography
    followers = profile.followers
    following = profile.followees
    print("Bio:", bio)
    print("Followers:", followers)
    print("Following:", following)
    
    # Step 1: Engagement Rate Calculation
    likes = []
    comments = []
    hashtags = []
    for post in profile.get_posts():
        likes.append(post.likes)
        comments.append(post.comments)
        hashtags.extend(post.caption_hashtags)
        if len(likes) >= 10:  # Limit to recent 10 posts
            break
    engagement_rate = calculate_engagement_rate(likes, comments, followers)
    print("Engagement Rate:", engagement_rate, "%")
    plot_engagement_rate(engagement_rate)  # Visualization for engagement rate

    # Step 2: Follower Growth Logging (requires daily run)
    with open('follower_growth.csv', 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([datetime.now(), followers])

    # Step 3: Keyword Extraction from Bio
    influencer_keywords = ["influencer", "content creator", "fashion", "travel", "beauty", "lifestyle", "fitness", "blogger", "photography", "fashionista", "makeup", "vlogger", "health", "sponsored", "collab", "advertising", "endorsements", "branding", "social media", "youtube", "tiktok", "digital marketer", "entrepreneur", "startup"]
    bio_terms = set(re.findall(r'\w+', bio.lower()))
    relevant_terms = bio_terms.intersection(influencer_keywords)
    print("Relevant terms in bio:", relevant_terms)
    plot_relevant_terms(relevant_terms)  # Visualization for relevant terms

    # Step 4: Hashtag Analysis
    print("Hashtags used in recent posts:", set(hashtags))

    # Step 5: Collaboration Rate (using Osintgram)
    tagged_accounts = run_osintgram_command(username, "tagged")
    if tagged_accounts:
        print("Tagged Accounts (potential collaborations):", tagged_accounts.split("\n")[:10])

    # Optional: External Data - Pricing and Rating (placeholder)
    print("For pricing and advanced ratings, refer to Hypeauditor or Social Blade for detailed reports.")
```

```python
# 7. Run the analysis (you will input the Instagram username when running the cell)
username = input("Enter the Instagram username for analysis: ")
instagram_osint_analysis(username)
```

```python
import instaloader
import re
import csv
from datetime import datetime
import subprocess
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from scipy.sparse import hstack

# Helper function to call Osintgram commands
def run_osintgram_command(username, command):
    try:
        output = subprocess.check_output(
            ["python3", "main.py", username, command], 
            cwd="C:/Users/Lavan/OneDrive/Desktop/College- SRH/Course/2nd Sem/OSINT/Practical-OSINT-Investigation_Instagram_2024-/LavanyaSrinivas06-Practical-OSINT-Investigation_Instagram_2024/Osintgram"
        )
        return output.decode("utf-8")
    except subprocess.CalledProcessError as e:
        print("Error running Osintgram command:", e)
        return None

# Function to calculate engagement rate
def calculate_engagement_rate(likes, comments, followers):
    avg_likes = sum(likes) / len(likes) if likes else 0
    avg_comments = sum(comments) / len(comments) if comments else 0
    return ((avg_likes + avg_comments) / followers) * 100

# Function to plot the engagement rate using Plotly
def plot_engagement_rate(engagement_rate):
    fig = go.Figure(go.Indicator(
        mode="gauge+number",
        value=engagement_rate,
        title={"text": "Engagement Rate (%)"},
        gauge={"axis": {"range": [0, 100]}, "bar": {"color": "blue"}}
    ))
    fig.show(renderer="browser")

# Function to plot the relevant terms in the bio
def plot_relevant_terms(relevant_terms):
    # Create a DataFrame with relevant terms and set their counts to 1
    df_terms = pd.DataFrame({'Term': list(relevant_terms), 'Count': [1] * len(relevant_terms)})

    fig = px.bar(
        df_terms,
        x='Term',
        y='Count',
        title="Bio Keyword Analysis",
        labels={'Term': 'Relevant Terms in Bio', 'Count': 'Frequency'}
    )
    fig.update_yaxes(visible=False)
    fig.show(renderer="browser")

# Function to preprocess the text (optional for cleaning)
def preprocess_text(text):
    # Add preprocessing steps like removing special characters, lowercasing, etc.
    return text.lower()

# Function to prepare the influencer data (CSV file)
def prepare_influencer_data():
    # Load the CSV file (adjust the path if needed)
    data = pd.read_csv(r"C:\Users\Lavan\OneDrive\Desktop\College- SRH\Course\2nd Sem\OSINT\Practical-OSINT-Investigation_Instagram_2024-\LavanyaSrinivas06-Practical-OSINT-Investigation_Instagram_2024\source\influencer_data.csv")
    
    # Clean the text in bio and posts
    data['bio_clean'] = data['bio'].apply(preprocess_text)
    data['posts_clean'] = data['posts'].apply(preprocess_text)
    
    # Extract features using TF-IDF vectorizer
    tfidf_vectorizer = TfidfVectorizer(max_features=1000)
    X_bio = tfidf_vectorizer.fit_transform(data['bio_clean'])
    X_posts = tfidf_vectorizer.transform(data['posts_clean'])
    
    # Combine bio and posts features
    X_tfidf = hstack([X_bio, X_posts])  # Combine bio and posts features
    
    # Labels (influencer type)
    y = data['influencer_type']
    
    return X_tfidf, y, tfidf_vectorizer

# Function to train the influencer classifier
def train_influencer_classifier(X_tfidf, y):
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.3, random_state=42)
    
    # Train a RandomForest classifier
    model = RandomForestClassifier()
    model.fit(X_train, y_train)
    
    # Evaluate the model
    accuracy = model.score(X_test, y_test)
    print(f"Model accuracy: {accuracy * 100:.2f}%")
    
    return model

# Main function for Instagram analysis
def instagram_osint_analysis(username):
    # Initialize Instaloader and fetch profile
    loader = instaloader.Instaloader()
    profile = instaloader.Profile.from_username(loader.context, username)

    # Basic profile info
    print(f"Analyzing @{username}")
    bio = profile.biography
    followers = profile.followers
    following = profile.followees
    print("Bio:", bio)
    print("Followers:", followers)
    print("Following:", following)
    
    # Step 1: Engagement Rate Calculation
    likes = []
    comments = []
    hashtags = []
    for post in profile.get_posts():
        likes.append(post.likes)
        comments.append(post.comments)
        hashtags.extend(post.caption_hashtags)
        if len(likes) >= 10:  # Limit to recent 10 posts
            break
    engagement_rate = calculate_engagement_rate(likes, comments, followers)
    print("Engagement Rate:", engagement_rate, "%")
    plot_engagement_rate(engagement_rate)  # Visualization for engagement rate

    # Step 2: Follower Growth Logging (requires daily run)
    with open('follower_growth.csv', 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([datetime.now(), followers])

    # Step 3: Keyword Extraction from Bio
    key_terms = ["travel", "fitness", "fashion", "lifestyle", "food", "blogger", "photography"]
    bio_terms = set(re.findall(r'\w+', bio.lower()))
    relevant_terms = bio_terms.intersection(key_terms)
    print("Relevant terms in bio:", relevant_terms)
    plot_relevant_terms(relevant_terms)  # Visualization for relevant terms

    # Step 4: Hashtag Analysis
    print("Hashtags used in recent posts:", set(hashtags))

    # Step 5: Collaboration Rate (using Osintgram)
    tagged_accounts = run_osintgram_command(username, "tagged")
    if tagged_accounts:
        print("Tagged Accounts (potential collaborations):", tagged_accounts.split("\n")[:10])

    # Optional: External Data - Pricing and Rating (placeholder)
    print("For pricing and advanced ratings, refer to Hypeauditor or Social Blade for detailed reports.")

    # Step 6: Predict Influencer Type (using pre-trained model)
    X_tfidf, y, tfidf_vectorizer = prepare_influencer_data()
    model = train_influencer_classifier(X_tfidf, y)

    # Prepare input data for the model based on profile bio and posts
    bio_cleaned = preprocess_text(bio)
    posts_cleaned = preprocess_text(' '.join([post.caption for post in profile.get_posts()]))
    input_data = tfidf_vectorizer.transform([bio_cleaned + ' ' + posts_cleaned])
    
    # Predict influencer type
    influencer_type = model.predict(input_data)
    print(f"Predicted Influencer Type: {influencer_type[0]}")

# Run the analysis
username = input("Enter the Instagram username for analysis: ")
instagram_osint_analysis(username)
```
